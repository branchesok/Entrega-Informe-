```python
labels_to_snake_case = lambda label: label.lower().replace('/',' ').replace(' ','_').replace('___','_').replace('__','_').replace('+','_').replace('-','_')
    
```

```python
snake_case_dict = {}


for label in df.columns[0:2]:
    snake_case_dict[label] = labels_to_snake_case(label)

for label in df.columns[2:]:
    snake_case_dict[label]= labels_to_snake_case(label)[1:-1]

snake_case_dict
```

```python
tesla_df = df.rename(columns= snake_case_dict)

tesla_df
```

```python
deaths_by_year =tesla_df.groupby('year')['deaths'].sum().sort_values(ascending=False)
deaths_by_year
```

```python
plot_deaths_by_year = deaths_by_year.plot(kind="barh",figsize=(10,10),title='Plot 2:Deaths by Year')
```

```python
deaths_by_country = tesla_df.groupby('country')['deaths'].sum().sort_values(ascending=False)
deaths_by_country
```

```python
plot_deaths_by_country = deaths_by_country.plot(kind="bar",figsize=(15,8),title='Plot 3 :Total deaths by Country')
```

```python
deaths_by_state = tesla_df.groupby('state')['deaths'].sum().sort_values(ascending=False)
deaths_by_state
```

```python
autopilot_deaths = tesla_df.groupby('autopilot_claimed')['deaths'].sum().sort_values(ascending=False)
```

```python
autopilot_deaths = autopilot_deaths.rename({True:"Yes",False:"No"})
autopilot_deaths
```
